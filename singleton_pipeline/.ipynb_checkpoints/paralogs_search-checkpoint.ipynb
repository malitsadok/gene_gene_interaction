```python
def load_file (path):
    chunk_size = 100000  # or whatever size you're using
    chunks = []

    for chunk in pd.read_csv(path, chunksize=chunk_size):
        chunks.append(chunk)

    df = pd.concat(chunks, ignore_index=True)
    return df 
```
