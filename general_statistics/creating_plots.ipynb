```python
import matplotlib.pyplot as plt

# Prepare data function
def prepare_mac(df):
    lof_mac = df.loc[df['n_lof'] > 0, 'n_lof']
    missense_mac = df.loc[df['n_missense'] > 0, 'n_missense']

    mac_counts_lof = lof_mac.value_counts().sort_index()
    mac_counts_missense = missense_mac.value_counts().sort_index()

    proportion_lof = mac_counts_lof / mac_counts_lof.sum()
    proportion_missense = mac_counts_missense / mac_counts_missense.sum()

    return mac_counts_lof, proportion_lof, mac_counts_missense, proportion_missense


# Prepare both datasets
mac_counts_lof_all, proportion_lof_all, mac_counts_mis_all, proportion_mis_all = prepare_mac(df_summary_all_snv)
mac_counts_lof_wb, proportion_lof_wb, mac_counts_mis_wb, proportion_mis_wb = prepare_mac(df_raw_data)

# Plot side by side
fig, axes = plt.subplots(1, 2, figsize=(16, 7), sharey=True)

# Left panel: All Population
axes[0].plot(mac_counts_mis_all.index, proportion_mis_all,
             drawstyle='steps-mid', color='lightgreen', linewidth=2.2, label='Missense')
axes[0].plot(mac_counts_lof_all.index, proportion_lof_all,
             drawstyle='steps-mid', color='lightblue', linewidth=2.2, label='LoF')
axes[0].set_xscale('log')
axes[0].set_yscale('log')
axes[0].set_xlabel('Allele Count', fontsize=16)
axes[0].set_ylabel('Proportion of Sites', fontsize=16)
axes[0].set_title("All Population", fontsize=20, fontweight='bold')
axes[0].tick_params(axis='both', which='major', labelsize=14)
axes[0].legend(fontsize=14)
# Add subplot label (a) OUTSIDE top-left corner
axes[0].annotate("(a)", xy=(0, 1.02), xycoords="axes fraction",
                 fontsize=18, fontweight="bold", ha="left", va="bottom")

# Right panel: White British
axes[1].plot(mac_counts_mis_wb.index, proportion_mis_wb,
             drawstyle='steps-mid', color='lightgreen', linewidth=2.2, label='Missense')
axes[1].plot(mac_counts_lof_wb.index, proportion_lof_wb,
             drawstyle='steps-mid', color='lightblue', linewidth=2.2, label='LoF')
axes[1].set_xscale('log')
axes[1].set_yscale('log')
axes[1].set_xlabel('Allele Count', fontsize=16)
axes[1].set_title("White British", fontsize=20, fontweight='bold')
axes[1].tick_params(axis='both', which='major', labelsize=14)
axes[1].legend(fontsize=14)
# Add subplot label (b) OUTSIDE top-left corner
axes[1].annotate("(b)", xy=(0, 1.02), xycoords="axes fraction",
                 fontsize=18, fontweight="bold", ha="left", va="bottom")

plt.tight_layout()

# Save one combined figure
plt.savefig("sfs_allpop_whitebritish.png", dpi=300)
plt.close()
```

```python
import matplotlib.pyplot as plt

# Function to prepare MAC distributions
def prepare_mac(df, lof_cutoff=234917, missense_cutoff=234917):
    # LoF: keep only those >0 and below cutoff
    lof_mac = df.loc[(df['n_lof'] > 0) & (df['n_lof'] < lof_cutoff), 'n_lof']

    # Missense: keep only those >0 and below cutoff
    missense_mac = df.loc[(df['n_missense'] > 0) & (df['n_missense'] < missense_cutoff), 'n_missense']

    # Counts of MAC values
    mac_counts_lof = lof_mac.value_counts().sort_index()
    mac_counts_missense = missense_mac.value_counts().sort_index()

    # Proportions
    proportion_lof = mac_counts_lof / mac_counts_lof.sum()
    proportion_missense = mac_counts_missense / mac_counts_missense.sum()

    # Optional: report how many SNVs were removed
    removed_lof = (df['n_lof'] >= lof_cutoff).sum()
    removed_mis = (df['n_missense'] >= missense_cutoff).sum()
    print(f"Removed {removed_lof} SNVs from LoF and {removed_mis} from missense")

    return mac_counts_lof, proportion_lof, mac_counts_missense, proportion_missense


# Prepare datasets
mac_counts_lof_all, proportion_lof_all, mac_counts_mis_all, proportion_mis_all = prepare_mac(df_summary_all_snv , 469834*0.4)
mac_counts_lof_wb, proportion_lof_wb, mac_counts_mis_wb, proportion_mis_wb = prepare_mac(df_raw_data, 431631*0.4)

# Plot side by side
fig, axes = plt.subplots(1, 2, figsize=(16, 7), sharey=True)

# Left panel: All Population
axes[0].plot(mac_counts_mis_all.index, proportion_mis_all,
             drawstyle='steps-mid', color='lightgreen', linewidth=2.2, label='Missense')
axes[0].plot(mac_counts_lof_all.index, proportion_lof_all,
             drawstyle='steps-mid', color='lightblue', linewidth=2.2, label='LoF')
axes[0].set_xscale('log')
axes[0].set_yscale('log')
axes[0].set_xlabel('Minor Allele Count', fontsize=16)
axes[0].set_ylabel('Proportion of Sites', fontsize=16)
axes[0].set_title("All Population", fontsize=20, fontweight='bold')
axes[0].tick_params(axis='both', which='major', labelsize=14)
axes[0].legend(fontsize=14)
# Add subplot label (a)
axes[0].annotate("(a)", xy=(0, 1.02), xycoords="axes fraction",
                 fontsize=18, fontweight="bold", ha="left", va="bottom")

# Right panel: White British
axes[1].plot(mac_counts_mis_wb.index, proportion_mis_wb,
             drawstyle='steps-mid', color='lightgreen', linewidth=2.2, label='Missense')
axes[1].plot(mac_counts_lof_wb.index, proportion_lof_wb,
             drawstyle='steps-mid', color='lightblue', linewidth=2.2, label='LoF')
axes[1].set_xscale('log')
axes[1].set_yscale('log')
axes[1].set_xlabel('Minor Allele Count', fontsize=16)
axes[1].set_title("White British", fontsize=20, fontweight='bold')
axes[1].tick_params(axis='both', which='major', labelsize=14)
axes[1].legend(fontsize=14)
# Add subplot label (b)
axes[1].annotate("(b)", xy=(0, 1.02), xycoords="axes fraction",
                 fontsize=18, fontweight="bold", ha="left", va="bottom")

plt.tight_layout()

# Save one combined figure
plt.savefig("sfs_allpop_whitebritish.png", dpi=300)
plt.close()
```

```python
import os
from glob import glob
import pandas as pd
from collections import defaultdict
option_lof = [ 'frameshift_variant', 'frameshift_variant&splice_region_variant', 'frameshift_variant&start_lost',
               'frameshift_variant&stop_gained', 'frameshift_variant&stop_gained&splice_region_variant',
               'frameshift_variant&stop_lost', 'start_lost', 'stop_gained', 'stop_gained&splice_region_variant',
               'stop_lost', 'start_lost&conservative_inframe_deletion',
               'start_lost&conservative_inframe_insertion', 'stop_gained&disruptive_inframe_deletion',
               'stop_lost&conservative_inframe_deletion', 'start_lost&splice_region_variant',
               'stop_gained&disruptive_inframe_deletion&splice_region_variant', 'stop_lost&splice_region_variant',
               'bidirectional_gene_fusion', 'stop_gained&conservative_inframe_insertion',
               'stop_gained&disruptive_inframe_insertion', 'start_lost&disruptive_inframe_deletion',
               'stop_lost&disruptive_inframe_deletion', 'gene_fusion', 'rare_amino_acid_variant',
               'stop_gained&conservative_inframe_insertion&splice_region_variant',
               'stop_gained&disruptive_inframe_insertion&splice_region_variant',
               'frameshift_variant&synonymous_variant',
               'start_lost&conservative_inframe_deletion&splice_region_variant',
               'frameshift_variant&stop_lost&splice_region_variant',
               'start_lost&disruptive_inframe_insertion',
               'frameshift_variant&missense_variant&splice_region_variant',
               'frameshift_variant&start_lost&splice_region_variant',
               'frameshift_variant&missense_variant',
               'stop_lost&conservative_inframe_deletion&splice_region_variant',
               'stop_gained&conservative_inframe_deletion',
               'start_lost&disruptive_inframe_deletion&splice_region_variant',
               'stop_lost&disruptive_inframe_deletion&splice_region_variant',
               'start_lost&conservative_inframe_insertion&splice_region_variant',
               'stop_lost&disruptive_inframe_insertion' ]

option_missense = ["missense_variant", "missense_variant&splice_region_variant",
                   "missense_variant&disruptive_inframe_insertion",
                   "missense_variant&conservative_inframe_insertion"]  # adjust if needed

def process_all_subfolders(root_dir):
    all_files = glob(os.path.join(root_dir, '**/*.csv'), recursive=True)

    results = defaultdict(lambda: {
        "all_genes": set(),
        "lof_genes": set(),
        "missense_genes": set(),
        "all_genes_homo": set(),
        "lof_genes_homo": set(),
        "missense_genes_homo": set()
    })

    for file_path in all_files:
        df = pd.read_csv(file_path)

        # all genes
        all_counts = df.groupby("Individual")["gene"].unique()
        for ind, genes in all_counts.items():
            results[ind]["all_genes"].update(genes)

        # LOF
        lof_counts = df[df["mutation_type"].isin(option_lof)].groupby("Individual")["gene"].unique()
        for ind, genes in lof_counts.items():
            results[ind]["lof_genes"].update(genes)

        # Missense
        missense_counts = df[df["mutation_type"].isin(option_missense)].groupby("Individual")["gene"].unique()
        for ind, genes in missense_counts.items():
            results[ind]["missense_genes"].update(genes)

        # Homozygous
        df_homo = df[df["count_Mut"] == 2]

        all_homo_counts = df_homo.groupby("Individual")["gene"].unique()
        for ind, genes in all_homo_counts.items():
            results[ind]["all_genes_homo"].update(genes)

        lof_homo_counts = df_homo[df_homo["mutation_type"].isin(option_lof)].groupby("Individual")["gene"].unique()
        for ind, genes in lof_homo_counts.items():
            results[ind]["lof_genes_homo"].update(genes)

        missense_homo_counts = df_homo[df_homo["mutation_type"].isin(option_missense)].groupby("Individual")["gene"].unique()
        for ind, genes in missense_homo_counts.items():
            results[ind]["missense_genes_homo"].update(genes)
        break

    # Convert results to DataFrame
    final_data = []
    for ind, vals in results.items():
        final_data.append({
            "Individual": ind,
            "all_genes": len(vals["all_genes"]),
            "lof_genes": len(vals["lof_genes"]),
            "missense_genes": len(vals["missense_genes"]),
            "all_genes_homo": len(vals["all_genes_homo"]),
            "lof_genes_homo": len(vals["lof_genes_homo"]),
            "missense_genes_homo": len(vals["missense_genes_homo"])
        })

    return pd.DataFrame(final_data)


if __name__ == "__main__":
    result =  process_all_subfolders("/sci/labs/orzuk/mali.tsadok/UKB/filter_data/lof")
    result.to_csv("/sci/labs/orzuk/mali.tsadok/UKB/data_per_individual.csv" , index= False)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# ===============================
# Load data
# ===============================
df_summary_all_snv = pd.read_csv(
    "C:/Users/maliz/thesa/UKbiobank/data/summary_all_SNV_new_method_with_1_percent.csv"
)
df_raw_data = pd.read_csv(
    "C:/Users/maliz/thesa/UKbiobank/data/summary_all_SNV_raw_data.csv"
)

# ===============================
# Plotting function
# ===============================
def plot_mac_proportions(ax, df, subtitle, N, highlight_singletons=True):
    """
    Plot site frequency spectrum for LoF and Missense variants.
    
    Parameters:
        ax: matplotlib axis
        df: dataframe containing 'n_lof' and 'n_missense'
        subtitle: str, panel title
        N: int, total number of alleles in the population
        highlight_singletons: bool, whether to highlight MAC=1
    """
    # Filter nonzero MACs
    lof_mac = df.loc[df['n_lof'] > 0, 'n_lof']
    missense_mac = df.loc[df['n_missense'] > 0, 'n_missense']

    # Count occurrences of each MAC
    mac_counts_lof = lof_mac.value_counts().sort_index()
    mac_counts_missense = missense_mac.value_counts().sort_index()

    # Convert counts to proportions
    proportion_lof = mac_counts_lof / mac_counts_lof.sum()
    proportion_missense = mac_counts_missense / mac_counts_missense.sum()

    # Main lines
    ax.plot(mac_counts_missense.index, proportion_missense,
            drawstyle='steps-mid', color='lightgreen', linewidth=2.2, label='Missense')
    ax.plot(mac_counts_lof.index, proportion_lof,
            drawstyle='steps-mid', color='lightblue', linewidth=2.2, label='LoF')

    # Highlight singletons
    if highlight_singletons:
        if 1 in mac_counts_lof.index:
            ax.plot(1, proportion_lof[1], 'o', color='blue', markersize=8, label='LoF Singletons')
        if 1 in mac_counts_missense.index:
            ax.plot(1, proportion_missense[1], 'o', color='green', markersize=8, label='Missense Singletons')

    # Vertical lines for 1% and 0.1% MAF
    ax.axvline(0.01 * N, color='red', linestyle='--', linewidth=1.5, label='1% MAF')
    ax.axvline(0.001 * N, color='orange', linestyle='--', linewidth=1.5, label='0.1% MAF')

    # Log-log scale
    ax.set_xscale('log')
    ax.set_yscale('log')

    # Labels & title
    ax.set_xlabel('Allele Count (MAC)', fontsize=14)
    ax.set_ylabel('Proportion of Sites', fontsize=14)
    ax.set_title(subtitle, fontsize=16, loc='left', fontweight='bold')

    # Legend
    ax.legend(fontsize=10)
    ax.tick_params(axis='both', which='major', labelsize=12)

# ===============================
# Population sizes (number of alleles)
# ===============================
# Assuming diploid individuals: total alleles = 2 * number of individuals
N_raw = df_raw_data.shape[0] * 2
N_summary = df_summary_all_snv.shape[0] * 2

# ===============================
# Create figure with two subplots
# ===============================
fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# Add main title for the figure
fig.suptitle("Site Frequency Spectrum of LoF and Missense Variants", fontsize=20, fontweight='bold')

# Plot each panel
plot_mac_proportions(axes[0], df_raw_data, "(a) Raw Data", N_raw)
plot_mac_proportions(axes[1], df_summary_all_snv, "(b) Filtered Summary", N_summary)

# Adjust layout to fit main title
plt.tight_layout(rect=[0, 0, 1, 0.96])

# Save figure
plt.savefig("C:/Users/maliz/thesa/UKbiobank/plots_for_thesis/Site_Frequency_Spectrum_LoF_Missense.png", dpi
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load data
df_raw_data = pd.read_csv("C:/Users/maliz/thesa/UKbiobank/data/summary_all_SNV_british.csv")

# Function for full SFS with vertical lines and MAF top axis
def plot_full_sfs(ax, df, subtitle, N):
    # Filter MAC > 0
    lof_mac = df.loc[df['n_lof'] > 0, 'n_lof']
    missense_mac = df.loc[df['n_missense'] > 0, 'n_missense']

    # Count allele occurrences
    lof_counts = lof_mac.value_counts().sort_index()
    missense_counts = missense_mac.value_counts().sort_index()

    # Plot SFS
    ax.plot(missense_counts.index, missense_counts / missense_counts.sum(),
            drawstyle='steps-mid', color='lightgreen', linewidth=2.2, label='Missense')
    ax.plot(lof_counts.index, lof_counts / lof_counts.sum(),
            drawstyle='steps-mid', color='lightblue', linewidth=2.2, label='LoF')

    # Vertical lines at 1% and 0.1% MAC
    ax.axvline(0.01 * N, color='red', linestyle='--', linewidth=1.5, label='1% MAF')
    ax.axvline(0.001 * N, color='orange', linestyle='--', linewidth=1.5, label='0.1% MAF')

    # Set log-log scale
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlabel('Allele Count (MAC)', fontsize=14)
    ax.set_ylabel('Proportion of Sites', fontsize=14)
    ax.set_title(subtitle, fontsize=16, fontweight='bold', loc='left')
    ax.legend(fontsize=10)
    ax.tick_params(axis='both', which='major', labelsize=12)

    # Add top x-axis showing MAF in %
    ax_top = ax.twiny()
    ax_top.set_xscale('log')
    ax_top.set_xlim(ax.get_xlim())
    # Example ticks at 0.1%, 1%, 5%
    maf_ticks = [0.001*N, 0.01*N, 0.05*N]
    ax_top.set_xticks(maf_ticks)
    ax_top.set_xticklabels(['0.1%', '1%', '5%'])
    ax_top.set_xlabel('Minor Allele Frequency (MAF)', fontsize=14)
    ax_top.tick_params(axis='x', labelsize=12)

# Function specifically for singleton plot (unchanged)
def plot_singletons(ax, df, subtitle):
    # Count singletons
    lof_singletons = (df['n_lof'] == 1).sum()
    missense_singletons = (df['n_missense'] == 1).sum()

    # Plot as bars
    ax.bar(0.8, lof_singletons, color='lightblue', width=0.4, label='LoF')
    ax.bar(1.2, missense_singletons, color='lightgreen', width=0.4, label='Missense')

    ax.set_xticks([1, 2])
    ax.set_xticklabels(['LoF', 'Missense'])
    ax.set_yscale('log')
    ax.set_ylabel('Number of SNVs', fontsize=14)
    ax.set_title(subtitle, fontsize=16, fontweight='bold', loc='left')
    ax.legend(fontsize=10)
    ax.tick_params(axis='both', which='major', labelsize=12)

# ======================
# Main plots
# ======================
N_raw = df_raw_data.shape[0] * 2

fig, axes = plt.subplots(1, 2, figsize=(16, 7))
fig.suptitle("Site Frequency Spectrum of LoF and Missense Variants", fontsize=20, fontweight='bold')

plot_full_sfs(axes[0], df_raw_data, "Full SFS", N_raw)
plot_singletons(axes[1], df_raw_data, "Singletons")

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.savefig("C:/Users/maliz/thesa/UKbiobank/plots_for_thesis/SFS_main.png", dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# -----------------------------
# Load data
# -----------------------------
df_raw_data = pd.read_csv("C:/Users/maliz/thesa/UKbiobank/data/summary_all_SNV_british.csv")

# Total number of individuals
N = 431631
total_alleles = 2 * N

# -----------------------------
# Prepare LoF and Missense counts
# -----------------------------
lof_mac = df_raw_data.loc[df_raw_data['n_lof'] > 0, 'n_lof']
missense_mac = df_raw_data.loc[df_raw_data['n_missense'] > 0, 'n_missense']

lof_counts = lof_mac.value_counts().sort_index()
missense_counts = missense_mac.value_counts().sort_index()

# Proportion of sites
lof_prop = lof_counts / lof_counts.sum()
missense_prop = missense_counts / missense_counts.sum()

# Allele count thresholds for 1% and 0.1% AF
maf_1pc = 0.01 * total_alleles
maf_0_1pc = 0.001 * total_alleles

# -----------------------------
# Plot 1: Log–log SFS
# -----------------------------
fig, ax = plt.subplots(figsize=(10, 7))

ax.plot(lof_prop.index, lof_prop, drawstyle='steps-mid',
        color='lightblue', linewidth=2.2, label='LoF')
ax.plot(missense_prop.index, missense_prop, drawstyle='steps-mid',
        color='lightgreen', linewidth=2.2, label='Missense')

# Vertical lines
ax.axvline(maf_1pc, color='red', linestyle='--', linewidth=1.5, label='1% AF')
ax.axvline(maf_0_1pc, color='orange', linestyle='--', linewidth=1.5, label='0.1% AF')

ax.set_xscale('log')
ax.set_yscale('log')

ax.set_xlabel('Allele count', fontsize=14)
ax.set_ylabel('Proportion of sites', fontsize=14)
ax.set_title('Site Frequency Spectrum of LoF and Missense Variants',
             fontsize=16, fontweight='bold')
ax.legend(fontsize=10)
ax.tick_params(axis='both', which='major', labelsize=12)
plt.savefig("C:/Users/maliz/thesa/UKbiobank/plots_for_thesis/Site Frequency Spectrum.png", dpi=300)

plt.tight_layout()
plt.show()

# -----------------------------
# Plot 2: Singletons — Linear scale bars
# -----------------------------
lof_singletons = (df_raw_data['n_lof'] == 1).sum()
missense_singletons = (df_raw_data['n_missense'] == 1).sum()

fig, ax = plt.subplots(figsize=(6, 6))

ax.bar(['LoF', 'Missense'],
       [lof_singletons, missense_singletons],
       color=['lightblue', 'lightgreen'],
       width=0.6)

ax.set_ylabel('Number of singleton sites', fontsize=14)
ax.set_title('Singleton LoF and Missense Variants', fontsize=16, fontweight='bold')
ax.tick_params(axis='both', which='major', labelsize=12)

# No log scale here
# ax.set_yscale('log')   <-- removed
plt.savefig("C:/Users/maliz/thesa/UKbiobank/plots_for_thesis/Singleton_distribution.png", dpi=300)

plt.tight_layout()
plt.show()
```
