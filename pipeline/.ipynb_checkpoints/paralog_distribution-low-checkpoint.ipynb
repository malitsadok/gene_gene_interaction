```python
import pandas as pd
def load_file(path, chunk_size=100_000):
    chunks = []
    for chunk in pd.read_csv(path, chunksize=chunk_size):
        chunks.append(chunk)
    return pd.concat(chunks, ignore_index=True)


df_lof_final_result = load_file("C:/Users/maliz/thesa/UKbiobank/python_and_R/pipeline/result/lof_final_result.csv")
```

```python
import os
import pandas as pd

# Set the folder path
folder_path = "C:/Users/maliz/thesa/UKbiobank/python_and_R/pipeline/chr17"

# List to store DataFrames
dfs = []

# Loop over all files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        file_path = os.path.join(folder_path, filename)
        df = pd.read_csv(file_path)
        dfs.append(df)

# Example: concatenate all dataframes into one
df_combined = pd.concat(dfs, ignore_index=True)
```
