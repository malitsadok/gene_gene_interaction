```python
import matplotlib.pyplot as plt
import numpy as np

# Select nonzero counts
lof_mac = df_summary_all_snv.loc[df_summary_all_snv['n_lof'] > 0, 'n_lof']
missense_mac = df_summary_all_snv.loc[df_summary_all_snv['n_missense'] > 0, 'n_missense']

# Count how many SNVs have each MAC
mac_counts_lof = lof_mac.value_counts().sort_index()
mac_counts_missense = missense_mac.value_counts().sort_index()

# Convert to proportions
proportion_lof = mac_counts_lof / mac_counts_lof.sum()
proportion_missense = mac_counts_missense / mac_counts_missense.sum()

# Plot
fig, ax = plt.subplots(figsize=(8, 6))

ax.plot(mac_counts_lof.index, proportion_lof, drawstyle='steps-mid', color='blue', label='LoF')
ax.plot(mac_counts_missense.index, proportion_missense, drawstyle='steps-mid', color='red', label='Missense')

# Linear scales
ax.set_xscale('linear')
ax.set_yscale('linear')

# Zoom limits
ax.set_xlim(1, 5)    # MAC 1 to 5
ax.set_ylim(0, 0.6)    # proportion 0 to 1

ax.set_xlabel('Allele Count')
ax.set_ylabel('Proportion of Sites')
ax.set_title('Site Frequency Spectrum of LoF and Missense Variants (Zoomed In)')
ax.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def smooth_middle(arr, window=3):
    # If array too short, just return it
    if len(arr) <= 2:
        return arr
    # Keep first and last values
    start, end = arr[0], arr[-1]
    # Smooth the middle values only
    middle = arr[1:-1]
    smoothed_middle = np.convolve(middle, np.ones(window)/window, mode='same')
    # Reconstruct full array
    return np.concatenate(([start], smoothed_middle, [end]))

# Original data
x_lof = mac_counts_lof.index.values
y_lof = proportion_lof.values

x_mis = mac_counts_missense.index.values
y_mis = proportion_missense.values

# Smooth middle part only
y_lof_smooth = smooth_middle(y_lof, window=3)
y_mis_smooth = smooth_middle(y_mis, window=3)

# Plot
fig, ax = plt.subplots(figsize=(8, 6))

ax.plot(x_lof, y_lof_smooth, color='blue', label='LoF Smoothed')
ax.plot(x_mis, y_mis_smooth, color='red', label='Missense Smoothed')

ax.set_xscale('linear')
ax.set_yscale('linear')
ax.set_xlim(1, 5)
ax.set_ylim(0, 1)

ax.set_xlabel('Allele Count')
ax.set_ylabel('Proportion of Sites')
ax.set_title('Smoothed Site Frequency Spectrum of LoF and Missense Variants')
ax.legend()

plt.tight_layout()
plt.show()
```

```python


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set Seaborn style
sns.set(style="whitegrid")

# Define input files and plot titles
files_info = [
    ("C:/Users/maliz/thesa/UKbiobank/python_and_R/pipeline/result/lof_final_result.csv", "expected_both_lof", "LoF"),
]

# Create figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("Expected Number of Individuals with Both Variants per Gene Pair\n(MAF ≤ 0.1%, Het and Hom Counted Equally)", fontsize=16)

for ax, (file_path, column, title) in zip(axes.flat, files_info):
    try:
        df = pd.read_csv(file_path, usecols=[column])
        df = df.dropna()
        df = df[df[column] <= 50]  # Limit x-axis to 50

        values = df[column]
        mean_val = values.mean()
        median_val = values.median()

        # Histogram + KDE
        sns.histplot(values, bins=50, kde=True, ax=ax,
                     color="skyblue", edgecolor="black",
                     log_scale=(False, True))

        ax.set_xlim(0, 30)
        ax.set_xlabel("Expected Individuals")
        ax.set_ylabel("Count (log scale)")
        ax.set_title(title)
        ax.legend()

    except Exception as e:
        ax.set_title(f"Error in {title}")
        print(f"Error with {title}: {e}")

# Add explanation as caption below the plot
caption = (
    "Each subplot shows the distribution of the expected number of individuals carrying both variants in a gene pair.\n"
    "The Minor Allele Frequency (MAF) cutoff is ≤ 0.1%.\n"
    "Both heterozygous and homozygous mutations are counted as one — i.e., if an individual is homozygous, they are still counted only once."
)
fig.text(0.5, -0.02, caption, ha='center', va='center', fontsize=11)

plt.tight_layout(rect=[0, 0.05, 1, 0.93])  # leave space for title and caption

# Save the figure
output_path = "/sci/labs/orzuk/mali.tsadok/UKB/all_data/expected_distribution_final_thesis_plot.png"
plt.show()
#plt.savefig(output_path, bbox_inches='tight', dpi=300)
#plt.close()

print(f"Saved thesis-ready plot to: {output_path}")
```
